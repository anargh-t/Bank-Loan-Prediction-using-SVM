```python
#input
x = df.iloc[:,:-1].values
x.ndim
x
```

```python
#output
y=df.iloc[:,-1].values
y.ndim
y
```

```python
#Training and Testing Data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
#Normalization
#StandardScaler
#z=(x-u)/s
#x : given data
#u : mean of training data
#s : standard deviation of training data

from sklearn.preprocessing import StandardScaler
scaler=StandardScaler() #object creation
scaler.fit(x_train)  #mean&sd of training data
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
x_train
```

```python
x_test
```

```python
#Model creation
#KNN
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=7)  #object create
knn.fit(x_train,y_train)  #model create
```

```python
#prediction
y_pred=knn.predict(x_test)
y_pred
```

```python
y_test
```

```python
#performance evaluation

from sklearn.metrics import confusion_matrix,accuracy_score
cmd=confusion_matrix(y_test,y_pred)
cmd
```

```python
#Accuracy socre
score=accuracy_score(y_test,y_pred)
score
```
